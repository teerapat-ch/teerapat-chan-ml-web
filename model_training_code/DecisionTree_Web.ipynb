```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
```

```python
data = pd.read_csv("/content/salary_dataset.csv")

df = pd.DataFrame({'Years': data['Years_of_Experience'], 'Salary': data['Salary']})
df = df.dropna()
df.plot(kind='scatter', x='Years', y='Salary', colorbar=False)
```

```python
years = np.array(df['Years']).reshape(-1, 1)
salary = np.array(df['Salary'])

years_list = list(years)
salary_list = list(salary)

dtree = DecisionTreeRegressor(random_state=0)
dtree.fit(years_list, salary_list)
```

```python
input = 26
prediction = dtree.predict(np.array([[input]]))

plt.scatter(years_list, salary_list, color='blue', label="Data Points")
plt.scatter(input, prediction[0], color='red', label="Prediction")
plt.text(x=input-1.2, y=prediction[0]-5000, s=f"Salary: {prediction[0]:.2f}", color='red')
plt.xlabel("Years")
plt.ylabel("Salary")
plt.title("Decision Tree Regression Prediction")
plt.show()
```

```python
import joblib

joblib.dump(dtree, 'dtree_model.pkl')
```
